```python
from skimage.io import imread, imsave
from skimage.color import rgb2gray, rgb2grey
inp_image = imread("./lena1.png")
img_gray= rgb2gray(inp_image)#gray can also be used
imsave('./lena_gray.png',img_gray)
imshow(img_gray)
```

## Binarization or thresholding

```python
from skimage.color import rgb2gray
from skimage.io import imread, imsave
from skimage.filters import threshold_otsu
from skimage import img_as_uint

inp_image = imread('lena1.png')
im_gray= rgb2gray(inp_image)
 
thresh= threshold_otsu(img_gray)
binary_thresh_img = img_gray > thresh

imsave('lena_thresh.png', img_as_uint(binary_thresh_img))
imshow(img_as_uint(binary_thresh_img))
```
